```python
# Install PIL (Pillow) and matplotlib in Google Colab
!pip install Pillow matplotlib

from google.colab import drive
drive.mount('/content/drive')

from PIL import Image, ImageDraw, ImageFont

# Load a background image (robotic design)
# Corrected file path: /content/drive/My Drive/roboticbg.jpg
background = Image.open('/content/drive/My Drive/roboticbg.jpg')

# Resize the background image to ID card size
background = background.resize((600, 400))

# Create a drawing object
draw = ImageDraw.Draw(background)

# Define the font size and font type (change to your path if needed)
font = ImageFont.truetype("/usr/share/fonts/truetype/dejavu/DejaVuSans-Bold.ttf", 25)

# Define text content for ID card
name = "Name: Najam Abbas"
father_name = "Father Name: Dilawar Abbas"
dob = "DOB: 30/05/1997"
class_name = "Class: Cloud Applied Generative AI"
time = "Time: Sunday 2 PM to 6 PM"

# Add text to image
draw.text((20, 100), name, fill="white", font=font)
draw.text((20, 150), father_name, fill="white", font=font)
draw.text((20, 200), dob, fill="white", font=font)
draw.text((20, 250), class_name, fill="white", font=font)
draw.text((20, 300), time, fill="white", font=font)

# Load user's profile picture
# Corrected file path: /content/drive/My Drive/student.jpg
profile_pic = Image.open('/content/drive/My Drive/student.jpg')

# Resize and position the profile picture
profile_pic = profile_pic.resize((100, 100))
background.paste(profile_pic, (450, 100))

# Save the final ID card image
background.save('/content/drive/My Drive/prifessional.id.card.jpg')

# Display the final ID card
background.show()
```
